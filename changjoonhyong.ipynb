```python
# import 해야하는 모듈들
import json
import requests
# 액세스 토큰 정보를 요청
# MariaDB에 아이디와 비밀번호를 입력하고 접근 권한을 받아오는 과정
# 후의 요청 헤더에 access_token 값을 넣어서 보내야 한다
url_auth = 'http://13.90.33.95:5000/pem/auth'
params = dict(
user = 'PEM',
password = 'T806T818T819'
)
resp = requests.post(url=url_auth, params=params)
data = json.loads(resp.text)
access_token = data['access_token']
refresh_token = data['refresh_token']
# access_token refresh
# access_token은 만료 기간이 짧으므로
# 토큰이 만료될 시에 refresh token을 통해서 재발급
url_refresh = 'http://13.90.33.95:5000/pem/refresh'
headers = dict(
Authorization = 'Bearer ' + refresh_token
)
resp = requests.post(url=url_refresh, headers=headers)
data = json.loads(resp.text)
access_token = data['access_token']
# 생성된 토큰을 바탕으로 Raw Data 정보 요청
# End Point 마지막의 이니셜을 통해서 사용자 데이터 구분
url_request = 'http://13.90.33.95:5000/pem/raw/KKS'
headers = dict(
Authorization = 'Bearer ' + access_token
)
resp = requests.post(url=url_request, headers=headers)
data = json.loads(resp.text)
print(data)
```
